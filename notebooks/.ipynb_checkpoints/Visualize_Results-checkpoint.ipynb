```python
def plot_annotated_image(img, rl_locations):
    # im = thresh_filtered_image
    im = img
    # Create figure and axes
    fig,ax = plt.subplots(1)

    # Display the image
    ax.imshow(im)

    # Create a Rectangle patch
    for entry in rl_locations:
        [tl_row, tl_col, br_row, br_col, conf] = entry
        kernel_col_size = br_col-tl_col
        kernel_row_size = br_row - tl_row
        rect = patches.Rectangle((tl_col, tl_row),kernel_col_size,kernel_row_size,linewidth=1,edgecolor='r',facecolor='none')
        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```

```python
count = 0
for filename_key in data_dict:
    datapath = '../../CS148/RedLights2011_Medium/'
    file = filename_key
    img = np.array(Image.open(datapath+file))
    rl_locations = data_dict[filename_key]
    plot_annotated_image(img, rl_locations)
    count += 1
#     if count > 3:
#         break
```
